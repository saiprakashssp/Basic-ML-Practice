```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('canadapercapitaincome.csv')
df
```

```python
%matplotlib inline
plt.xlabel('Year')
plt.ylabel('PerCapitaIncome_US_Dollars')
plt.scatter(df.Year,df.PerCapitaIncome_US_Dollars,color='red',marker='+')
```

```python
new_df = df.drop('PerCapitaIncome_US_Dollars',axis='columns')
new_df
```

```python
PerCapitaIncome_US_Dollars=df.PerCapitaIncome_US_Dollars
PerCapitaIncome_US_Dollars
```

```python
# Create linear regression object
reg = linear_model.LinearRegression()
reg.fit(new_df,PerCapitaIncome_US_Dollars)
```

(1) Predict Per Capita Income of a Canada in the year 2021

```python
reg.predict([[2021]])
```

Generate CSV file with list of Per Capita Income predictions

```python
year_df = pd.read_csv("years.csv")
year_df.head(5)
```

```python
p = reg.predict(year_df)
p
```

```python
year_df['PerCapitaIncome_US_Dollars']=p
year_df
```

```python
year_df.to_csv("prediction.csv")
```
